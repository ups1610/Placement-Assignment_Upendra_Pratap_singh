```python
#import useful libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#load data from Question-4 as being extracated
df = pd.read_csv("meteorite.csv")
df.drop(['Unnamed: 0'],axis=1,inplace=True)
df.head()
```

```python
#pre-process the data to extract the useful information
#extract year from the year column
def year_extraction(df,year):
    for i in range(0,len(df['year'])):
        year.append(str(df['year'][i]).split("-")[0])
year = []
year_extraction(df,year)

#add year to dataframe
df['get_year']=year
```

***Insights to be drawn -***
##### ● Get all the Earth meteorites that fell before the year 2000
##### ● Get all the earth meteorites co-ordinates who fell before the year 1970
##### ● Assuming that the mass of the earth meteorites was in kg, get all those whose mass was more than 10000kg

```python
# Get all the Earth meteorites that fell before the year 2000

# Filter meteorites that fell before the year 2000
filtered_meteorites = df[df["get_year"]< '2000']

# Count the number of meteorites by year
meteorite_count_by_year = filtered_meteorites["get_year"].value_counts().sort_index()

# Plotting the number of meteorites by year
plt.figure(figsize=(10, 6))
plt.plot(meteorite_count_by_year.index, meteorite_count_by_year.values)
plt.xlabel("Year")
plt.ylabel("Number of Meteorites")
plt.title("Number of Earth Meteorites (Before Year 2000)")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Get all the earth meteorites co-ordinates who fell before the year 1970

# Filter meteorites that fell before the year 1970
filtered_meteorites = df[df['get_year'] < '1970']

# Extract latitude and longitude coordinates
coordinates = filtered_meteorites[['reclat', 'reclong']]

# # Plotting the coordinates on a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(coordinates['reclong'], coordinates['reclat'], s=10)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earth Meteorites that Fell before 1970')
plt.tight_layout()
plt.show()

```

```python
#Assuming that the mass of the earth meteorites was in kg, get all those whose mass was morethan 10000kg
filter_data = df[df['mass']>10000.0]

#plot the graph
plt.figure(figsize=(10, 6))
plt.bar(filter_data["name"], filter_data["mass"],width=2)
plt.xlabel("Meteorite")
plt.ylabel("Mass greater than 10000 Kg")
plt.title("earth meteorites whose mass greater than 10000 Kg")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
